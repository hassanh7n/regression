```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
```

```python
corr_with_profit = df[['Profit', 'Sales', 'Quantity', 'Discount']].corr()

corr_with_profit['Profit'].sort_values(ascending=False)
```

```python
df_encoded = pd.get_dummies(
    df[['Category', 'Region', 'Profit']], drop_first=True
)

df_encoded.corr()['Profit'].sort_values(ascending=False)
```

```python
X = df[['Discount', 'Quantity']]
y = df['Profit']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
```

```python
plt.plot(y_test, color='blue', label='Test')
plt.plot(y_train_pred, color='red', label='Predictions')
plt.legend()
plt.show()
```
